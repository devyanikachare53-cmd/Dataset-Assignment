Q10. (Mini Project - Complete Preprocessing Pipeline)
Create a complete preprocessing pipeline for the Ford Car Dataset:
1. Load and clean the data (handle missing & duplicate values).
2. Perform EDA (histograms + count plots + heatmap).
3. Identify input and output features.
4. Encode categorical variables.
5. Scale the numeric features.

```python

# Q10: Complete Preprocessing Pipeline

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# 1. Data Cleaning

print("Missing Values:")
print(df.isnull().sum())

print("\nDuplicate Rows:", df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()

print("Duplicate Rows After Removal:", df.duplicated().sum())

# 2. EDA - Histograms

plt.figure(figsize=(12,8))

plt.subplot(2,3,1)
plt.hist(df['price'], bins=30)
plt.title("Price")

plt.subplot(2,3,2)
plt.hist(df['mileage'], bins=30)
plt.title("Mileage")

plt.subplot(2,3,3)
plt.hist(df['year'], bins=30)
plt.title("Year")

plt.subplot(2,3,4)
plt.hist(df['engineSize'], bins=30)
plt.title("Engine Size")

plt.subplot(2,3,5)
plt.hist(df['mpg'], bins=30)
plt.title("MPG")

plt.tight_layout()
plt.show()

# 3. EDA - Count Plots

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.countplot(x='fuelType', data=df)
plt.title("Fuel Type")

plt.subplot(1,3,2)
sns.countplot(x='transmission', data=df)
plt.title("Transmission")

plt.subplot(1,3,3)
sns.countplot(y='model', data=df)
plt.title("Model")

plt.tight_layout()
plt.show()

# 4. Correlation Heatmap

plt.figure(figsize=(8,6))

corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")
plt.show()

# 5. Input and Output Features

X = df.drop('price', axis=1)
y = df['price']

print("\nInput Features:")
print(X.columns)

print("\nOutput Feature:")
print(y.name)

# 6. One Hot Encoding

print("\nBefore Encoding:")
print(df[['model','fuelType','transmission']].head())

df_encoded = pd.get_dummies(
    df,
    columns=['model','fuelType','transmission']
)

print("\nAfter Encoding:")
print(df_encoded.head())

# 7. Feature Scaling

scaler = StandardScaler()

num_cols = ['year','mileage','tax','mpg','engineSize']

df_encoded[num_cols] = scaler.fit_transform(
    df_encoded[num_cols]
)

print("\nFirst 5 Rows of Scaled Data:")
print(df_encoded.head())

# Summary

# 1. Checked missing values.
# 2. Removed duplicate rows.
# 3. Created histograms for numeric features.
# 4. Created count plots for categorical features.
# 5. Created correlation heatmap.
# 6. Selected input features (X) and output feature (price).
# 7. Applied One Hot Encoding.
# 8. Applied Standard Scaling.
# 9. Dataset is ready for machine learning.
```
