```python
sns.histplot(data['IMDB_Rating'])         #Numeric data distribution column plot
plt.title("Distribution of IMDb Ratings")
plt.xlabel("IMDb Rating")
plt.ylabel("Number of Movies")
```

```python
data["IMDB_Rating"].describe()
```

```python
sns.histplot(data['Meta_score'])
plt.title('Distribution of Meta Score ')
plt.xlabel("Meta Score")
plt.ylabel("Count")
```

```python
data['Meta_score'].describe()
```

```python
sns.histplot(data['No_of_Votes'])
plt.title('Distribution of No of Vote')
plt.xlabel('No of vote')
plt.ylabel('Count')
```

```python
# sns.countplot(y=data['Certificate'])
sns.heatmap(data.corr(numeric_only=True),annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
```

```python
certificate_count = data['Certificate'].value_counts()
plt.pie(certificate_count,
        labels=certificate_count.index,
        autopct='%1.1f%%')
plt.title("Movie Certificate Distribution")
```

```python
genre_count = data['Genre'].value_counts().head(5)

plt.figure(figsize=(8,8))
plt.pie(genre_count,
        labels=genre_count.index,
        autopct='%1.1f%%')          #distribute percentage on pie chart
plt.title("Top 5 Genres")
plt.show()
```

```python
sns.boxplot(x=data['IMDB_Rating'])
plt.title("Box Plot of IMDb Ratings")
```
