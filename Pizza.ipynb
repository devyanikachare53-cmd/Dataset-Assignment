```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.isnull().sum()
```

```python
duplicate_count = df.duplicated().sum()
df = df.drop_duplicates()
print("Duplicates removed.")

print("New shape of the df:", df.shape)
```

```python
print("Distribution plots for all numerical columns ")
# hist_plot=plt.hist(df["RestingBP"])

plt.subplot(2,2,1)
sns.histplot(df["order_id"])
plt.title("Order id distribution")
plt.xlabel("order_id")
plt.ylabel("Frequency")

plt.subplot(2,2,2)
sns.histplot(df["quantity"])
plt.title("Qunatity Distiution for Pizza")
plt.xlabel("quantity")
plt.ylabel("Frequency")

plt.subplot(2,2,3)
sns.histplot(df["total_price"])
plt.title("Total Price rate Distribution")
plt.xlabel("Total price")
plt.ylabel("Frequency")



plt.subplot(2,2,4)
sns.histplot(df["order_details_id"])
plt.title(" Order Details for Pizza customer distribution")
plt.xlabel("Order_details")
plt.ylabel("Frequency")

plt.tight_layout()
```

```python
plt.plot(df["order_id"])
plt.title("Order id ")
plt.xlabel("Rate")
plt.ylabel("frequency")
```

```python
sns.histplot(df["pizza_id"])
plt.xlabel("pizza id")
```

```python
top10 = df.groupby("pizza_name")["quantity"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(7,5))
top10.plot(kind="bar")
plt.title("Top 10 Selling Pizzas")
plt.xlabel("Pizza Name")
plt.ylabel("Quantity Sold")

```

```python
category_sales = df.groupby("pizza_category")["quantity"].sum()

plt.figure(figsize=(6,6))
plt.pie(category_sales,
        labels=category_sales.index,
        autopct="%1.1f%%")

plt.title("Pizza Category Sales")
```

```python
size_revenue = df.groupby("pizza_size")["total_price"].sum()

plt.figure(figsize=(7,5))
size_revenue.plot(kind="bar")

plt.title("Revenue by Pizza Size")
plt.xlabel("Pizza Size")
plt.ylabel("Revenue")
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["total_price"], bins=30)

plt.title("Distribution of Total Price")
```

```python
plt.figure(figsize=(8,5))

sns.countplot(data=df,x="pizza_category")

plt.title("Pizza Category Count")
```

```python
numeric_df = df.select_dtypes(include=["int64","float64"])

plt.figure(figsize=(8,6))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
```
